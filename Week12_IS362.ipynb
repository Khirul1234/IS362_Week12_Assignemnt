```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# the code below will read and display the csv data
MP = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data')
MP.head(10)
```

```python
MP = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data', sep=',', header=None, usecols=[0,5,21], names=["Edible/Poisonous","Odor","Population"])
MP.head(10)
```

```python
MP.replace(to_replace={'Edible/Poisonous':{'p': 0, 'e':1}}, inplace=True)
MP.replace(to_replace={'Odor':{'a':2, 'l':3, 'c':4, 'y':5, 'f':6, 'm':7, 'n':8, 'p':9, 's':10}}, inplace=True)
MP.replace(to_replace={'Population':{'a':11, 'c':12, 'n':13, 's':14, 'v':15, 'y':16}}, inplace=True)
MP.head(10)
```

```python
sns.set(style='whitegrid', context='poster')
g = sns.lmplot(hue='Edible/Poisonous', x='Population', y='Odor', data=MP, size = 20)
g.set_axis_labels('Population', 'Odor')
sns
```


Conclusion:

Odor is the most helpful criteria in determining what mushrooms are edible/poisonous based on the data.
